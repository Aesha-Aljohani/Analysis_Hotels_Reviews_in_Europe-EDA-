# Analysis the most Nationality staying in hotels in Europe(MVP)

### In this graph,  it shows the analysis of the most nationalities staying in hotels in europe

```python
text_nationality = " ".join(Reviewer_Nationality for Reviewer_Nationality in df_Hotel.Reviewer_Nationality)
```

```python
# Creating word_cloud with text as argument in .generate() method

word_cloud2 = WordCloud(collocations = False, background_color = 'white').generate(text_nationality)

# Display the generated Word Cloud
fig, ax = plt.subplots(figsize=(10, 5))

plt.imshow(word_cloud2, interpolation='bilinear')

plt.axis("off")
plt.show()
```
